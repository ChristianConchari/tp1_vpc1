# Importaci칩n de librerias

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import os
```

# Carga y visualizaci칩n de im치genes

Visualizamos las im치genes `img1_tp.png` y `img2_tp.png` en escala de grises.

```python
# path to images
path_img1 = 'resources/img1_tp.png'
path_img2 = 'resources/img2_tp.png'
# read images
img1 = cv2.imread(path_img1, cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread(path_img2, cv2.IMREAD_GRAYSCALE)
# show the images
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(img1, cmap='gray')
axs[0].set_title('Image 1')
axs[0].axis('off')
axs[1].imshow(img2, cmap='gray')
axs[1].set_title('Image 2')
axs[1].axis('off')
```

# Visualizaci칩n de histogramas

```python
# calculate histograms
img1_hist, bins1 = np.histogram(img1.ravel(), bins=50, range=[0, 256])
img2_hist, bins2  = np.histogram(img2.ravel(), bins=50, range=[0, 256])
# show histograms
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].hist(img1.ravel(),256,[0,256])
axs[0].set_title('Histogram of Image 1')
axs[0].set_xlabel('Pixel value')
axs[0].set_ylabel('Frequency')
axs[1].hist(img2.ravel(),256,[0,256])
axs[1].set_title('Histogram of Image 2')
axs[1].set_xlabel('Pixel value')
axs[1].set_ylabel('Frequency')

plt.show()
```

```python
# calculate the difference between the histograms
difference = np.sum(np.abs(img1_hist - img2_hist))

# print the difference
print('Difference between the histograms:', difference)
```

#### An치lisis:

Se observa que la imagen `img1_tp.png` y la imagen `img2_tp.png` tienen la misma distribuci칩n de intensidades. Si bien visualmente se observa que ambas im치genes son diferentes, el histograma de ambas im치genes es el mismo. Esto es posible debido a que el histograma solo representa la distribuci칩n de intensidades de la imagen, y no el ordenamiento espacial de los p칤xeles.

Sobre tomar los histogramas como **features** para entrenar un modelo de **clasificaci칩n/detecci칩n**, lo encontrado en este ejercicio sugiere que no, ya que dos im치genes diferentes pueden tener el mismo histograma. Si tomaramos este caso por ejemplo, un modelo no ser칤a capaz de distinguir entre ambas im치genes, aunque evidentemente son diferentes.
